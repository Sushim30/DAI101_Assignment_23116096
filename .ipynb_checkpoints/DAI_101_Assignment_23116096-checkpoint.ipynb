```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('BMIData.csv')
# i am taking data from the BMIData.csv , it was taken from kaggle.com
```

```python
print(df.head())
```

```python
df.info() #here i checked for null enteries present in the data 
```

```python
print(df.describe()) # it gives the main statistical information of all the columns in my dataframe
```

```python
print(df.isnull().sum())  
```

```python
#trying to fill in the empty spaces in each column
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for col in num_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled missing values in numeric column '{col}' with median value: {median_val}")
```

```python
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
```

```python
print(df.isnull().sum()) #checked for any null entries if left
```

```python
num_duplicates = df.duplicated().sum()   #checking for double enteries of the same info in the datframe
print(f"\nNumber of duplicate records: {num_duplicates}")
df.drop_duplicates(inplace=True)
print(f"Shape after duplicate removal: {df.shape}")
```

```python
#i am defining a method to remove the outliers
def remove_outliers_IQR(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
```

```python
for col in num_cols:
    n_before = df.shape[0]
    df = remove_outliers_IQR(df, col)
    n_after = df.shape[0]
    print(f"Removed {n_before - n_after} outliers from column '{col}'.")
```

```python
# Convert text in categorical columns to lowercase and remove extra spaces.
for col in cat_cols:
    df[col] = df[col].str.lower().str.strip()
    print(f"Standardized text in column '{col}'.")
```

```python
#finding median , mode , standard deviation and other things of the dataset
print(df.describe(include='all'))
```

```python
#doing univariate analysis on my dataset
```

```python
for col in cat_cols:
    print(f"\nFrequency distribution for '{col}':")
    print(df[col].value_counts())
```

```python
for col in num_cols:    #plotting the histogram plot for better visualisation of the data
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig(f'hist_{col}.png')
    plt.show()
```

```python
for col in num_cols:    #making the box plots and as we know that outliers have been already removed, therefore there should be no points after the extreme of the whiskers
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.savefig(f'box_{col}.png')
    plt.show()
```

```python
corr_matrix = df[num_cols].corr()   
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.tight_layout()
plt.savefig('correlation_matrix.png')
plt.show()
#made the heatmap for the correlation matrix of two columns at a time 
```

```python
sns.pairplot(df[num_cols])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.savefig('pairplot.png')
plt.show()
```

```python
# merging the data for males and females to showcase the differenes in various columns 
if 'Sex' in df.columns:
    grouped_means = df.groupby('Sex')[num_cols].mean()
    print("\nGrouped Mean Values by Gender:")
    print(grouped_means)
```
